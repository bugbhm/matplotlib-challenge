```python
#dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
#read csv files
mouse_drug = pd.read_csv("mouse_drug_data.csv")
clinical_trial = pd.read_csv("clinicaltrial_data.csv")
```

```python
#merge on Mouse ID
drug_data = pd.merge(clinical_trial, mouse_drug, on = 'Mouse ID', how = 'left')
drug_data.head(5)
```

# Tumor Volume After Treatement

```python
#group by drug and find average tumor volume
drugs_grouped = drug_data.groupby(['Drug', 'Timepoint'])
avg_tumor_vol = drugs_grouped.mean()['Tumor Volume (mm3)']

#reset index
avg_tumor_vol = avg_tumor_vol.reset_index()
avg_tumor_vol
```

```python
# find standard error of tumor vol
se_tumor_vol = drugs_grouped.sem()["Tumor Volume (mm3)"]
se_tumor_change = pd.DataFrame(se_tumor_vol)

se_tumor_change = se_tumor_change.reset_index()
se_tumor_change.head()
```

```python
#munging
tumor_vol_df = avg_tumor_vol.pivot(index= 'Timepoint', columns='Drug', values='Tumor Volume (mm3)')
tumor_vol_df = pd.DataFrame(tumor_vol_df)

tumor_vol_df.head()
```

```python
# plot change in tumor vol
fig, ax = plt.subplots(figsize=(12,5))
x_axis = tumor_vol_df.index.values
markers = ['o','o','o','o','o','o','o','o','o','o']
colors = ['red', 'blue', 'green', 'yellow', 'magenta', 'cyan', 'orange', 'limegreen', 'grey', 'black']

#loop
i = 0
for drug in tumor_vol_df.columns:
    y_axis = tumor_vol_df[drug].values
    standard_errors = se_tumor_change[se_tumor_change.Drug == drug]['Tumor Volume (mm3)'].values
    ax.errorbar(x_axis, y_axis, standard_errors, label=drug, linestyle='solid', marker=markers[i], color=colors[i])
    i += 1

plt.title("Tumor Changes After Drug")
plt.xlabel("Timepoints")
plt.ylabel("Tumor Volume")
plt.legend(loc='upperleft')
plt.grid()

# save chart
plt.savefig('tumor_changes.png')
```

# Percent Change/Summary

```python
# find percent change in vol for each drug
avg_tumor_df = avg_tumor_vol.drop(['Timepoint'], axis=1)
drug_grouped_df = avg_tumor_df.groupby(['Drug'])

# % = last value - first value / first value * 100
percent_changes = (drug_grouped_df['Tumor Volume (mm3)'].last() - drug_grouped_df['Tumor Volume (mm3)'].first()) /\
                drug_grouped_df['Tumor Volume (mm3)'].first() * 100
percent_changes
```

```python
#find percent changes
percent_changes = percent_changes.sort_values(ascending=True)

failing_drug = percent_changes[percent_changes < 0].sort_values(ascending=True)
passing_drug = percent_changes[percent_changes > 0].sort_values(ascending=True)

# label percentages of changes
plt.figure(figsize=(14,4))

negative_axis = plt.bar(failing_drug.index.values, failing_drug.values, color='green', width=1)
positive_axis = plt.bar(passing_drug.index.values, passing_drug.values, color='red', width=1)

# label chart
plt.title("Tumor Change Throughout Treatment")
plt.ylabel("Percent Tumor Volume Change")
plt.xlim(-0.7,len(percent_changes)-0.3)
plt.ylim(-30,70)
plt.grid()

# find negative changes
negative_changes = negative_axis.patches
for change in negative_changes:
    y_value = change.get_height()
    x_value =change.get_x() + change.get_width() / 2
    percentY = str(int(y_value)) + '%'
    plt.text(x_value-0.1, -5, percentY, color='w')

# find postive changes
positive_changes = positive_axis.patches
for change in positive_changes:
    y_value = change.get_height()
    x_value = change.get_x() + change.get_width() / 2
    percentY = str(int(y_value)) + '%'
    plt.text(x_value-0.1, 5, percentY, color='w')

# Save the Figure
plt.savefig("total_tumor_vol_change.png")
```
